To test for the stationarity of a time series data, you can use various statistical tests. One common method is the Augmented Dickey-Fuller (ADF) test.

This function adf_test takes a time series data as input and prints out the ADF statistic, p-value, and critical values. You would interpret the results as follows:

If the p-value is less than a significance level (e.g., 0.05), you reject the null hypothesis of the ADF test, indicating that the time series is stationary.
If the p-value is greater than the significance level, you fail to reject the null hypothesis, indicating that the time series is non-stationary.


The critical values in the Augmented Dickey-Fuller (ADF) test represent the threshold values that determine whether the null hypothesis of non-stationarity is rejected or not. These critical values depend on the significance level chosen for the test (usually 1%, 5%, and 10%) and the sample size.

When you perform the ADF test, you typically compare the ADF statistic with these critical values to determine the stationarity of the time series data. Here's how to interpret the results:

If the ADF statistic is less than the critical value at a chosen significance level, you reject the null hypothesis of non-stationarity and conclude that the time series is stationary.
If the ADF statistic is greater than the critical value, you fail to reject the null hypothesis, indicating that the time series is non-stationary.
The critical values are provided in the ADF test results, and they are usually printed out along with the ADF statistic and the p-value. They are calculated based on the type of ADF test performed (e.g., with or without trend), the significance level, and the sample size.

The Augmented Dickey-Fuller (ADF) statistic is a test statistic used in the Augmented Dickey-Fuller test, which is a statistical test for testing the stationarity of a time series. The ADF test is commonly used in econometrics and time series analysis.

The ADF statistic is computed as part of the ADF test procedure and measures how strongly a time series is affected by a unit root, which indicates non-stationarity. The presence of a unit root implies that the time series is non-stationary, meaning its mean and variance change over time.

In the ADF test, the null hypothesis is that the time series has a unit root and is non-stationary. The ADF statistic is compared to critical values to determine whether to reject the null hypothesis. If the ADF statistic is less than the critical value, the null hypothesis is rejected, indicating that the time series is stationary. If the ADF statistic is greater than the critical value, the null hypothesis is not rejected, suggesting that the time series is non-stationary.

The ADF statistic is calculated based on the regression coefficient estimated in a regression model that includes lagged differences of the time series. The exact formula for the ADF statistic depends on the specific version of the ADF test being used (e.g., with or without a trend term).

The Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) are two important tools used in time series analysis to understand the autocorrelation structure of a time series data and to identify the appropriate order of autoregressive and moving average terms in a time series model, such as ARIMA.

1. Autocorrelation Function (ACF):
   - The ACF measures the correlation between a time series and its lagged values at different lags.
   - It quantifies the relationship between observations at different time points.
   - ACF is used to identify the presence of autocorrelation in a time series.
   - In the ACF plot, significant autocorrelation at various lags indicates the presence of patterns or seasonality in the time series data.
   - The ACF plot typically decays over time, suggesting that the autocorrelation decreases as the lag increases.

2. Partial Autocorrelation Function (PACF):
   - The PACF measures the correlation between two observations in a time series while controlling for the effect of other observations between them.
   - It provides the correlation between two observations at a specific lag, excluding the influence of the intervening observations.
   - PACF is used to determine the order of the autoregressive (AR) component in an ARIMA model.
   - In the PACF plot, significant partial autocorrelation at a particular lag indicates the presence of a direct relationship between observations at that lag, suggesting the appropriate order of the AR term in the ARIMA model.

The ACF and PACF plots are visual tools commonly used in time series analysis to identify the order of autoregressive and moving average terms in ARIMA models. By examining these plots, analysts can determine the appropriate values of p (AR order) and q (MA order) in the ARIMA(p,d,q) model. The ACF plot helps in determining the value of q, while the PACF plot helps in determining the value of p.

In summary, the ACF and PACF plots are essential for diagnosing the autocorrelation structure of a time series data and selecting the appropriate parameters for building ARIMA models, thereby enabling accurate forecasting of future values.

The nlags parameter specifies the number of lags to include in the ACF and PACF plots. Adjust the value of nlags according to the length of your time series data and the desired level of detail in the plots.

The Autocorrelation Function (ACF) plot displays the correlation between a time series and its lagged values. Here's how to interpret the ACF plot:

1. **Lag on the x-axis:** The horizontal axis of the ACF plot represents the number of lags, which are the time intervals between the current observation and the lagged observations.

2. **Autocorrelation on the y-axis:** The vertical axis of the ACF plot represents the autocorrelation coefficients. These coefficients measure the correlation between the time series and its lagged values at different lags.

3. **Interpretation of autocorrelation values:**
   - Autocorrelation values close to 1 indicate a strong positive correlation between the time series and its lagged values at that lag.
   - Autocorrelation values close to -1 indicate a strong negative correlation between the time series and its lagged values at that lag.
   - Autocorrelation values close to 0 indicate weak or no correlation between the time series and its lagged values at that lag.

4. **Significance levels:** In some cases, horizontal lines are drawn on the ACF plot to indicate the significance levels for autocorrelation. These significance levels help identify statistically significant correlations. Typically, a correlation is considered statistically significant if it falls outside the shaded region, which is often based on a confidence interval.

5. **Pattern identification:** Analyzing the pattern of autocorrelation values in the ACF plot can provide insights into the underlying structure of the time series data. For example:
   - If autocorrelation values decay exponentially as the lag increases, it suggests a geometric decay pattern, indicating a stationary time series.
   - If autocorrelation values alternate in sign or display a sinusoidal pattern, it suggests a non-stationary or cyclical time series.

Overall, the ACF plot helps identify potential autoregressive (AR) terms in an autoregressive integrated moving average (ARIMA) model and provides insights into the temporal dependence structure of the time series data.

If the autocorrelation function (ACF) plot indicates a cyclical pattern in the time series data, it suggests that the data may exhibit seasonality or periodic fluctuations. In such cases, the traditional ARIMA model may not be sufficient because it is designed to handle stationary time series data with no seasonality.

However, you can still use an extension of the ARIMA model called the Seasonal ARIMA (SARIMA) model. SARIMA incorporates additional seasonal parameters to capture the seasonality in the data. It includes seasonal autoregressive (SAR) and seasonal moving average (SMA) terms along with the non-seasonal ARIMA terms.

To determine the appropriate seasonal parameters for the SARIMA model, you can analyze the seasonal autocorrelation function (SACF) plot, which is similar to the ACF plot but focuses on seasonal lags.

Here are the general steps to handle a cyclical time series using SARIMA:

1. **Identify seasonality:** Examine the ACF and PACF plots to identify any seasonal patterns in the data. Look for significant spikes at seasonal lags in the ACF and PACF plots.

2. **Differencing:** If the data is non-stationary, apply differencing to make it stationary. You may need to apply seasonal differencing if there is seasonality.

3. **Model selection:** Choose appropriate ARIMA and seasonal parameters (p, d, q) and (P, D, Q) based on the ACF, PACF, and SACF plots. You can use methods like grid search or auto-correlation-based information criteria (e.g., AIC, BIC) to select the best model.

4. **Model estimation:** Fit the SARIMA model to the differenced data using the selected parameters.

5. **Diagnostic checking:** Evaluate the residuals of the fitted model to ensure they are white noise. You can use residual diagnostics plots (e.g., ACF of residuals, histogram of residuals) for this purpose.

6. **Forecasting:** Once you have a satisfactory model, use it to forecast future values of the time series.

By following these steps and using the SARIMA model, you can effectively handle cyclical time series data and make accurate forecasts.

The Partial Autocorrelation Function (PACF) plot is a diagnostic tool used in time series analysis to identify the appropriate order of the autoregressive (AR) component in an Autoregressive Integrated Moving Average (ARIMA) model. Here's how to interpret PACF plots:

1. **Definition of PACF:** The PACF at lag k represents the correlation between the series and its lagged values, with the effects of shorter lags removed.

2. **Interpretation of PACF Values:** Each bar in the PACF plot represents the partial correlation between the series and its lagged values at different lags. The height of the bar indicates the strength of the correlation. The PACF values are interpreted in the following way:
   - If the PACF value at lag k is significantly different from zero (i.e., it exceeds the significance threshold), it suggests a direct relationship between the series at time t and the series at time t-k, after removing the effects of shorter lags.
   - PACF values that exceed the significance threshold at lag k and are also higher than those at shorter lags indicate a potential AR component of order k in the ARIMA model.

3. **Identification of AR Component:** In general, for identifying the order of the AR component:
   - Look for significant spikes in the PACF plot at different lags.
   - The number of significant spikes indicates the order of the AR component. For example, if there is a significant spike at lag 1 but no significant spikes at higher lags, it suggests an AR(1) component.

4. **Interpretation Example:** Suppose you observe a significant spike at lag 1 and another significant spike at lag 2 in the PACF plot, with no significant spikes at higher lags. This pattern suggests that there is a strong correlation between the series at time t and the series at time t-1 (lag 1), and there is also a correlation between the series at time t and the series at time t-2 (lag 2). This indicates a potential AR(2) component in the ARIMA model.

5. **Model Selection:** Use the information from both the ACF and PACF plots to select appropriate orders for the AR, MA, and possibly seasonal components of the ARIMA model.

In summary, the PACF plot helps in identifying the appropriate order of the AR component in the ARIMA model by showing significant spikes at different lags, which indicate direct relationships between the series at different time points after accounting for shorter lags.

If the Augmented Dickey-Fuller (ADF) statistic is lower than the significant values (indicating stationarity), but the Autocorrelation Function (ACF) shows a cyclical pattern with autocorrelation, it suggests that there may still be some residual autocorrelation in the data. This residual autocorrelation might not be adequately captured by the current model.

In such cases, it's essential to consider whether the cyclical pattern in the ACF is due to seasonality or other underlying patterns in the data. If seasonality is present, you may need to consider a Seasonal Autoregressive Integrated Moving Average (SARIMA) model, which can capture both the autocorrelation and seasonality in the data.

Alternatively, if the cyclical pattern in the ACF is not due to seasonality but rather some other underlying pattern, you may need to explore more complex models or incorporate additional features or exogenous variables into your model to better capture the data's behavior.

In summary, even if the ADF test indicates stationarity, it's essential to carefully examine the ACF and consider whether any remaining patterns need to be addressed in the modeling process. This may involve choosing a more appropriate model or incorporating additional features or variables.

**Wavelet ANN model**

```python
# Wavelet ANN model for rainfall and lake levels
import pandas as pd
import numpy as np
import pywt
from google.colab import drive
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import itertools

# Function to perform wavelet decomposition
def wavelet_transform(data, wavelet='db1', level=1):
    coeffs = pywt.wavedec(data, wavelet, level=level)
    return np.concatenate(coeffs)

# Load the data
drive.mount('/content/drive/', force_remount=True)
file_path = '/content/drive/MyDrive/dataset/Lake_Tana_1996_2006_monthly.csv'

df = pd.read_csv(file_path)
#df['Date'] = pd.to_datetime(df['Date'], format='%m/%Y')

# Split the data into training and testing sets
split_index = int(0.2 * len(df))
train_data = df[:split_index]
test_data = df[split_index:]

# Prepare the training and testing data
rain_X_train = train_data[['Mean Rainfall']].values
level_y_train = train_data['Lake Level'].values
rain_X_test = test_data[['Mean Rainfall']].values
level_y_test = test_data['Lake Level'].values
date_test = test_data['Date'].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
rain_X_train_scaled = scaler.fit_transform(rain_X_train)
rain_X_test_scaled = scaler.transform(rain_X_test)

# Perform wavelet transform and extract features
wavelet_level = 2  # Adjust as needed
rain_X_train_wavelet = np.apply_along_axis(wavelet_transform, 1, rain_X_train_scaled, level=wavelet_level)
rain_X_test_wavelet = np.apply_along_axis(wavelet_transform, 1, rain_X_test_scaled, level=wavelet_level)

# Build the ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(rain_X_train_wavelet.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
#model.fit(rain_X_train_wavelet, flow_y_train, epochs=100, batch_size=64)
model.fit(rain_X_train_wavelet, level_y_train, epochs=100, batch_size=32)

# Make predictions
level_y_pred = model.predict(rain_X_test_wavelet)
level_y_pred = scaler.inverse_transform(level_y_pred)

# Evaluate the model
score = r2_score(level_y_test, level_y_pred)
mse = mean_squared_error(level_y_test, level_y_pred)
print('R-squared score:', score)
print('Mean Squared Error:', mse)

# Prepare to calculate Nash-Sutcliffe Efficiency (NSE)and PBIAS

# Observed and predicted values
observed_values = level_y_test
predicted_values = level_y_pred

# Calculate mean of observed values
observed_mean = np.mean(observed_values)

# Calculate Nash-Sutcliffe Efficiency (NSE)
nse = 1 - (np.sum((observed_values - predicted_values) ** 2) / np.sum((observed_values - observed_mean) ** 2))

# Calculate Percent Bias (PBIAS)
pbias = (np.sum(observed_values - predicted_values) / np.sum(observed_values)) * 100

print("Nash-Sutcliffe Efficiency (NSE):", nse)
print("Percent Bias (PBIAS):", pbias)

# Plot outputs
plt.scatter(rain_X_test, level_y_test, color='black', label='Observed Flow')
plt.plot(rain_X_test, level_y_pred, color='blue', linewidth=3, label='Predicted Flow')
plt.xlabel('Rainfall')
plt.ylabel('Lake Level')
plt.title('Wavelet-ANN Forecasting Lake Levels')
plt.legend()
plt.show()

# Plot time series of observed flow and predicted flow
plt.figure(figsize=(12, 6))
plt.plot(date_test, level_y_test, label='Observed Lake Levels', color='blue')
plt.plot(date_test, level_y_pred, label='Predicted Lake Levels', linestyle='--', color='red')
#plt.plot(flow_y_test, label='Observed Flow', color='blue')
#plt.plot(flow_y_pred, label='Predicted Flow', linestyle='--', color='red')
plt.xlabel('Date')
plt.ylabel('Lake Level (masl)')
plt.title('W-ANN: Time Series of Observed Lake Levels and Predicted Lake Levels')
plt.legend()

#Set range for y-axis
plt.ylim(1780, 1790)

plt.grid(True)
plt.tight_layout()
plt.show()

# Write model results to CSV file
#results_df = pd.DataFrame({'Date': Date['Date'], 'Rainfall': rain_X_test.ravel(),
#                           'Observed Lake Levels': level_y_test, 'Predicted Lake Levels': level_y_pred.ravel()})
#results_df.to_csv('/content/drive/MyDrive/dataset/output/Tana_pred_wavelet_ANN_flow.csv', index=False)
#print("File saved successfully!")
```

**SARIMA**

```python
#SARIMA model to fit a timeseries of rainfall to lake levels
#Change % split of training, and p.d.q values
#IMPORT LIBARIES
import pandas as pd # pandas for data manipulation
import numpy as np # numpy for numerical operations
from google.colab import drive # used in Google Colab notebooks to mount Google Drive
import matplotlib.pyplot as plt # for plotting a graph.
import matplotlib.dates as mdates # for plotting dates on graphs
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import r2_score, mean_squared_error

import itertools
import warnings
warnings.filterwarnings("ignore")

# NAME AND STUDENT ID
print('Andrew Lowe')
print ('ID 1223456')

# LOAD DATA CSV FILES

# Mount Google Drive
drive.mount('/content/drive/', force_remount=True)

# Load the temperature and observed flow data in CSV file from Google Drive
file_path = '/content/drive/MyDrive/dataset/Lake_Tana_1996_2006_monthly.csv'

# Read climate data from CSV file and create a DataFrame
df = pd.read_csv(file_path)

# Convert 'Date' column to datetime object
#df['Date'] = pd.to_datetime(df['Date'], format='%m/%Y')


# SPLIT DATA INTO TRAINING AND TEST SETS AND PREPARE DATA SETS

# Split the data into training and testing sets (30% training, 70% testing then 50%).
#split_index = int(0.3 * len(df)) # % of the data for training
split_index = int(0.5 * len(df)) # % of the data for training

# The df[:split_index] is called DataFrame slicing, where the train_data = df[:split_index] creates a new DataFrame called train_data containing the first split_index rows of the original DataFrame df
train_data = df[:split_index] # Training data contains the first % of the data
test_data = df[split_index:] # Testing data contains the remaining % of the data

#print(train_data.columns)

# Prepare the training and testing data from CSV file
rain_X_train = train_data[['Mean Rainfall', 'Mean ETO']].values
level_y_train = train_data['Lake Level'].values
rain_X_test = test_data[['Mean Rainfall', 'Mean ETO']].values
level_y_test = test_data['Lake Level'].values
date_test = test_data ['Date']

# DEFINE p, d, q, P, D, Q, and m PARAMETER VALUES AND MAKE PREDICTION OF FLOW USING SARIMA ON TESTING DATA

# Define range of values for p, d, q, P, D, Q, and m
p_values = range(0, 5)  # Autoregression order
d_values = range(0, 3)  # Differencing order
q_values = range(0, 5)  # Moving average order
P_values = range(0, 2)  # Seasonal autoregression order
D_values = range(0, 2)  # Seasonal differencing order
Q_values = range(0, 2)  # Seasonal moving average order

m = 30  # Seasonal period (assuming 30 days per month)

best_aic = float("inf")  # Initialize best AIC
best_order = None
best_seasonal_order = None

# Iterate over all combinations of p, d, q, P, D, Q values
for p, d, q, P, D, Q in itertools.product(p_values, d_values, q_values, P_values, D_values, Q_values):
    order = (p, d, q)
    seasonal_order = (P, D, Q, m)
    try:
        # Fit SARIMA model
        model = SARIMAX(level_y_train, exog=rain_X_train, order=order, seasonal_order=seasonal_order,
                        enforce_stationarity=False, enforce_invertibility=False)
        sarima_results = model.fit()

        # Calculate AIC
        aic = sarima_results.aic

        # Update best AIC, order, and seasonal_order if current model has lower AIC
        if aic < best_aic:
            best_aic = aic
            best_order = order
            best_seasonal_order = seasonal_order
    except:
        continue

print("Best AIC:", best_aic)
print("Best Order (p, d, q):", best_order)
print("Best Seasonal Order (P, D, Q, m):", best_seasonal_order)

# Fit SARIMA model with best order and seasonal order
best_model = SARIMAX(level_y_train, exog=rain_X_train, order=best_order, seasonal_order=best_seasonal_order,
                     enforce_stationarity=False, enforce_invertibility=False)
best_sarima_results = best_model.fit()

# Make predictions using SARIMA on testing data
level_y_pred = best_sarima_results.forecast(steps=len(rain_X_test), exog=rain_X_test)
#level_y_pred = best_sarima_results.forecast(steps=len(rain_X_test))

# EVALUATE MODEL PERFORMACE

# EVALUATE THE ARIMA MODEL PERFORMANCE

# Calculate the R-squared score
score = r2_score(level_y_test, level_y_pred)
print('R-squared score:', score)

# EVALUATE THE PERFORMANCE OF THE GLACIER RUNOFF MODEL

# Evaluate runoff model results (1) with simple r2 and mean standard error (mse)

# Calculate evaluation metrics for total predicted flow
r2 = r2_score(level_y_test, level_y_pred)
mse = mean_squared_error(level_y_test, level_y_pred)

# Print evaluation metrics for total predicted flow
print("Evaluation Metrics for Total Predicted Flow:")
print(f"R2: {r2:.2f}")
print(f"MSE: {mse:.2f}")

# Evaluate runof model results (2) using Nash-Sutcliffe Efficiency (NSE)and PBIAS

# Prepare to calculate Nash-Sutcliffe Efficiency (NSE)and PBIAS

# Observed and predicted values
observed_values = level_y_test
predicted_values = level_y_pred

# Calculate mean of observed values
observed_mean = np.mean(observed_values)

# Calculate Nash-Sutcliffe Efficiency (NSE)
nse = 1 - (np.sum((observed_values - predicted_values) ** 2) / np.sum((observed_values - observed_mean) ** 2))

# Calculate Percent Bias (PBIAS)
pbias = (np.sum(observed_values - predicted_values) / np.sum(observed_values)) * 100

print("Nash-Sutcliffe Efficiency (NSE):", nse)
print("Percent Bias (PBIAS):", pbias)


# CREATE SCATTER PLOT OF TEMPERATURE AND PREDICTED FLOW FOR TEST DATA

# Plot outputs
#plt.scatter(rain_X_test, level_y_test, color="black")
#plt.plot(rain_X_test, level_y_pred, color="blue", linewidth=3)

#plt.xlabel('Rainfall')
#plt.ylabel('Predicted Lake Levels')
#plt.title('SARIMA: Rainfall vs Lake Levels')

plt.show()


# CREATE TIME SERIES PLOT OF OBSERVED FLOW and PREDICTED FLOW FOR TEST PERIOD

# Plot time series of observed flow and predicted flow
plt.figure(figsize=(12, 6))
plt.plot(date_test, level_y_test, label='Observed Lake Levels', color='blue')
plt.plot(date_test, level_y_pred, label='Predicted Lake Levels', linestyle='--', color='red')
plt.xlabel('Date')
plt.ylabel('Lake Level (masl)')
plt.title('SARIMA: Time Series of Observed Lake Level and Predicted Lake Level')
plt.legend()

#Set range for y-axis
plt.ylim(1780, 1790)

# Set x-axis major and minor intervals
#plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))  # Major interval: 1 day
#plt.gca().xaxis.set_minor_locator(mdates.HourLocator(interval=12))  # Minor interval: 6 hours
#plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))  # Format the major ticks as 'DD-MM-YYYY'

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Plot GRID
plt.grid(True)
plt.tight_layout()

plt.show()


# WRITE MODEL RESULTS TO CSV FILE

# Write the predictions to a file on Google Drive and confirm file is saved

# Convert multi-dimensional arrays to one-dimensional arrays if necessary
rain_X_test = rain_X_test.ravel()  # Ensure temp_X_test is one-dimensional
level_y_pred = level_y_pred.ravel()  # Ensure flow_y_pred is one-dimensional

# Create DataFrame with one-dimensional arrays
#results_df = pd.DataFrame({'Date': Date,'Rainfall': rain_X_test, 'Observed Lake Levels': level_y_test, 'Predicted Lake Level': level_y_pred})

# Write the DataFrame to a CSV file
#results_df.to_csv('/content/drive/MyDrive/dataset/output/Tana_pred_SARIMA_ML_level.csv', index=False)

#print("File saved successfully!")
```
